```python
from google.colab import drive
drive.mount('/content/drive')

# ==============================
# 1. РҳРјРҝРҫСҖСӮ РұРёРұР»РёРҫСӮРөРә
# ==============================
import os
import pandas as pd

# ==============================
# 2. РҹСғСӮСҢ Рә РҙР°РҪРҪСӢРј
# ==============================
DATA_PATH = "/content/drive/MyDrive/Colab Notebooks/city"

# ==============================
# 3. РҹРҫР»СғСҮРөРҪРёРө СҒРҝРёСҒРәР° parquet-С„Р°Р№Р»РҫРІ
# ==============================
files = sorted([f for f in os.listdir(DATA_PATH) if f.endswith(".parquet")])

print(f"РқР°Р№РҙРөРҪРҫ С„Р°Р№Р»РҫРІ: {len(files)}")
print("РҹСҖРёРјРөСҖ С„Р°Р№Р»РҫРІ:", files[:3])

# ==============================
# 4. Р—Р°РіСҖСғР·РәР° Рё РҝСҖРөРҙРҫРұСҖР°РұРҫСӮРәР° РҙР°РҪРҪСӢС…
# ==============================
dfs = []

for file in files:
    path = os.path.join(DATA_PATH, file)

    df = pd.read_parquet(path)

    # ------------------------------
    # РіРҫСҖРҫРҙ РёР· РёРјРөРҪРё С„Р°Р№Р»Р°
    # ------------------------------
    city = file.split("_")[0]
    df["city"] = city

    # ------------------------------
    # РІРҫСҒСҒСӮР°РҪРҫРІР»РөРҪРёРө РІСҖРөРјРөРҪРё
    # ------------------------------
    df = df.copy()
    df["time"] = pd.to_datetime(df.index)

    # ------------------------------
    # РҙРҫРҝРҫР»РҪРёСӮРөР»СҢРҪСӢРө РҝСҖРёР·РҪР°РәРё РІСҖРөРјРөРҪРё
    # ------------------------------
    df["year"] = df["time"].dt.year
    df["month"] = df["time"].dt.month
    df["day"] = df["time"].dt.day
    df["weekday"] = df["time"].dt.weekday
    df["hour"] = df["time"].dt.hour

    dfs.append(df)

# ==============================
# 5. РһРұСҠРөРҙРёРҪРөРҪРёРө РҙР°РҪРҪСӢС… РҝРҫ РІСҒРөРј РіРҫСҖРҫРҙР°Рј
# ==============================
df_all = pd.concat(dfs, ignore_index=True)

# ==============================
# 6. Р‘Р°Р·РҫРІР°СҸ РҝСҖРҫРІРөСҖРәР° РҙР°РҪРҪСӢС…
# ==============================
print("\n===== Р”РҗРқРқР«Р• Р—РҗР“Р РЈР–Р•РқР« =====")
print("Р Р°Р·РјРөСҖ РҙР°СӮР°СҒРөСӮР°:", df_all.shape)

print("\nР Р°СҒРҝСҖРөРҙРөР»РөРҪРёРө РҝРҫ РіРҫСҖРҫРҙР°Рј:")
print(df_all["city"].value_counts())

print("\nРЎРҝРёСҒРҫРә РҝСҖРёР·РҪР°РәРҫРІ:")
print(df_all.columns)

print("\nРҹСҖРёРјРөСҖ РҙР°РҪРҪСӢС…:")
display(df_all.head())

# ==============================
# 7. Р Р°Р·РҙРөР»РөРҪРёРө РҙР°РҪРҪСӢС… РҝРҫ РіРҫСҖРҫРҙР°Рј (РҙР»СҸ РҝРҫСҒР»РөРҙСғСҺСүРөРіРҫ РјРҫРҙРөР»РёСҖРҫРІР°РҪРёСҸ)
# ==============================
city_dfs = {
    city: df_all[df_all["city"] == city].copy()
    for city in df_all["city"].unique()
}

print("\nР“РҫСҖРҫРҙР°, РҝРҫРҙРіРҫСӮРҫРІР»РөРҪРҪСӢРө РҙР»СҸ РјРҫРҙРөР»РёСҖРҫРІР°РҪРёСҸ:", list(city_dfs.keys()))
```

Р‘СӢР»Р° РІСӢРҝРҫР»РҪРөРҪР° Р°РіСҖРөРіР°СҶРёСҸ РјРҪРҫРіРҫР»РөСӮРҪРёС… РјРөСӮРөРҫРҙР°РҪРҪСӢС… РҝРҫ 6 РіРҫСҖРҫРҙР°Рј, СҒС„РҫСҖРјРёСҖРҫРІР°РҪ РөРҙРёРҪСӢР№ РІСҖРөРјРөРҪРҪРҫР№ РҙР°СӮР°СҒРөСӮ СҒ СҒРҫС…СҖР°РҪРөРҪРёРөРј РІСҖРөРјРөРҪРҪРҫР№ СҒСӮСҖСғРәСӮСғСҖСӢ Рё РҙРҫРұР°РІР»РөРҪРёРөРј РәР°Р»РөРҪРҙР°СҖРҪСӢС… РҝСҖРёР·РҪР°РәРҫРІ.

```python
# ==============================
# 2.1 EDA
# ==============================

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

from statsmodels.tsa.stattools import adfuller

sns.set(style="whitegrid")

# ==============================
# 0. Р‘РҗР—РһР’Р«Р• РҹР РһР’Р•Р РҡРҳ
# ==============================
print("Shape:", df_all.shape)

print("\nРҹСҖРҫРҝСғСҒРәРё:")
print(df_all.isnull().sum())

# ==============================
# 1. РўР Р•РқР”Р« (Р’Р Р•РңР•РқРқР«Р• Р РҜР”Р«)
# ==============================
plt.figure(figsize=(14,6))

for city in df_all["city"].unique():
    temp = df_all[df_all["city"] == city].sort_values("time")
    plt.plot(temp["time"], temp["temperature_2m"], label=city, alpha=0.7)

plt.title("РўРөРјРҝРөСҖР°СӮСғСҖР° РҝРҫ РіРҫСҖРҫРҙР°Рј (РІСҖРөРјРөРҪРҪСӢРө СҖСҸРҙСӢ)")
plt.legend()
plt.show()

# ==============================
# 2. РЎР•Р—РһРқРқРһРЎРўР¬ (Р“РһР”РһР’РҗРҜ)
# ==============================
monthly = df_all.groupby(["city", "month"])["temperature_2m"].mean().reset_index()

plt.figure(figsize=(12,6))
sns.lineplot(data=monthly, x="month", y="temperature_2m", hue="city")
plt.title("Р“РҫРҙРҫРІР°СҸ СҒРөР·РҫРҪРҪРҫСҒСӮСҢ СӮРөРјРҝРөСҖР°СӮСғСҖСӢ")
plt.show()

# ==============================
# 3. РЎР•Р—РһРқРқРһРЎРўР¬ (РЎРЈРўРһР§РқРҗРҜ)
# ==============================
hourly = df_all.groupby(["city", "hour"])["temperature_2m"].mean().reset_index()

plt.figure(figsize=(12,6))
sns.lineplot(data=hourly, x="hour", y="temperature_2m", hue="city")
plt.title("РЎСғСӮРҫСҮРҪР°СҸ СҒРөР·РҫРҪРҪРҫСҒСӮСҢ СӮРөРјРҝРөСҖР°СӮСғСҖСӢ")
plt.show()

# ==============================
# 4. Р РҗРЎРҹР Р•Р”Р•РӣР•РқРҳРҜ (HIST + KDE)
# ==============================
features = ['temperature_2m','relative_humidity_2m','wind_speed_10m','surface_pressure']

for col in features:
    plt.figure(figsize=(6,4))
    sns.histplot(df_all[col], kde=True)
    plt.title(f"Р Р°СҒРҝСҖРөРҙРөР»РөРҪРёРө: {col}")
    plt.show()

# ==============================
# 5. BOXPLOT (Р РҗРЎРҹР Р•Р”Р•РӣР•РқРҳР• РҹРһ Р“РһР РһР”РҗРң)
# ==============================
plt.figure(figsize=(12,6))
sns.boxplot(data=df_all, x="city", y="temperature_2m")
plt.title("РўРөРјРҝРөСҖР°СӮСғСҖР° РҝРҫ РіРҫСҖРҫРҙР°Рј")
plt.xticks(rotation=30)
plt.show()

plt.figure(figsize=(12,6))
sns.boxplot(data=df_all, x="city", y="precipitation")
plt.title("РһСҒР°РҙРәРё РҝРҫ РіРҫСҖРҫРҙР°Рј")
plt.xticks(rotation=30)
plt.show()

# ==============================
# 6. Q-Q PLOT (РқРһР РңРҗРӣР¬РқРһРЎРўР¬)
# ==============================
for col in features:
    plt.figure(figsize=(5,5))
    stats.probplot(df_all[col].sample(5000), dist="norm", plot=plt)
    plt.title(f"Q-Q plot: {col}")
    plt.show()

# ==============================
# 7. РҡРһР Р Р•РӣРҜРҰРҳРҳ
# ==============================
corr = df_all.drop(columns=["city", "time"]).corr()

plt.figure(figsize=(10,8))
sns.heatmap(corr, cmap="coolwarm")
plt.title("РҡРҫСҖСҖРөР»СҸСҶРёРҫРҪРҪР°СҸ РјР°СӮСҖРёСҶР°")
plt.show()

# ==============================
# 8. РҡР РһРЎРЎ-Р“РһР РһР”РЎРҡРһРҷ РҗРқРҗРӣРҳР—
# ==============================
annual = df_all.groupby("city").agg({
    "temperature_2m": "mean",
    "precipitation": "sum"
}).reset_index()

print("\nРҡР»РёРјР°СӮРёСҮРөСҒРәРёРө РҝСҖРҫС„РёР»Рё РіРҫСҖРҫРҙРҫРІ:")
display(annual)

# ==============================
# 9. РҗРқРһРңРҗРӣРҳРҳ (РҹР РһРЎРўРһРҷ РҹРһР”РҘРһР”)
# ==============================
df_all["temp_zscore"] = (
    (df_all["temperature_2m"] - df_all["temperature_2m"].mean()) /
    df_all["temperature_2m"].std()
)

anomalies = df_all[np.abs(df_all["temp_zscore"]) > 3]

print("\nРҡРҫР»РёСҮРөСҒСӮРІРҫ Р°РҪРҫРјР°Р»РёР№:", len(anomalies))

# ==============================
# 10. РЎРўРҗРҰРҳРһРқРҗР РқРһРЎРўР¬ (ADF)
# ==============================
print("\n===== ADF TEST =====")

for city in df_all["city"].unique():
    series = df_all[df_all["city"] == city]["temperature_2m"].dropna()
    result = adfuller(series[:5000])

    print(f"\n{city}")
    print(f"ADF stat: {result[0]:.4f}")
    print(f"p-value: {result[1]:.6f}")

    if result[1] < 0.05:
        print("=> РЎСӮР°СҶРёРҫРҪР°СҖРҪСӢР№")
    else:
        print("=> РқР• СҒСӮР°СҶРёРҫРҪР°СҖРҪСӢР№")

# ==============================
# 11. Р”РҳРӨРӨР•Р Р•РқРҰРҳР РһР’РҗРқРҳР•
# ==============================
df_all["temp_diff"] = df_all.groupby("city")["temperature_2m"].diff()

print("\nADF РҝРҫСҒР»Рө РҙРёС„С„РөСҖРөРҪСҶРёСҖРҫРІР°РҪРёСҸ (РҝСҖРёРјРөСҖ РңРҫСҒРәРІР°):")
series = df_all[df_all["city"]=="РңРҫСҒРәРІР°"]["temp_diff"].dropna()
print(adfuller(series[:5000]))

# ==============================
# 12. РҹР РһРҹРЈРЎРҡРҳ (Р’РҳР—РЈРҗРӣРҳР—РҗРҰРҳРҜ)
# ==============================
missing = df_all.isnull().mean() * 100

plt.figure(figsize=(10,4))
missing.sort_values().plot(kind="bar")
plt.title("РҹСҖРҫСҶРөРҪСӮ РҝСҖРҫРҝСғСҒРәРҫРІ")
plt.show()
```

Р’СӢРІРҫРҙСӢ РҝРҫ EDA (2.1)
РҗРҪР°Р»РёР· РјРөСӮРөРҫРҙР°РҪРҪСӢС… РҝРҫРәР°Р·Р°Р»:

РўРөРјРҝРөСҖР°СӮСғСҖР° РІРҫ РІСҒРөС… РіРҫСҖРҫРҙР°С… РҙРөРјРҫРҪСҒСӮСҖРёСҖСғРөСӮ РІСӢСҖР°Р¶РөРҪРҪСғСҺ РіРҫРҙРҫРІСғСҺ СҒРөР·РҫРҪРҪРҫСҒСӮСҢ, Р° СӮР°РәР¶Рө СғРјРөСҖРөРҪРҪСғСҺ СҒСғСӮРҫСҮРҪСғСҺ РҙРёРҪР°РјРёРәСғ

Р Р°СҒРҝСҖРөРҙРөР»РөРҪРёСҸ РҝСҖРёР·РҪР°РәРҫРІ РҫСӮРәР»РҫРҪСҸСҺСӮСҒСҸ РҫСӮ РҪРҫСҖРјР°Р»СҢРҪРҫРіРҫ Р·Р°РәРҫРҪР°, СҮСӮРҫ РҝРҫРҙСӮРІРөСҖР¶РҙРөРҪРҫ Q-Q РіСҖР°С„РёРәР°РјРё

РңРөР¶РҙСғ РіРҫСҖРҫРҙР°РјРё РҪР°РұР»СҺРҙР°СҺСӮСҒСҸ Р·РҪР°СҮРёРјСӢРө СҖР°Р·Р»РёСҮРёСҸ:

СҺР¶РҪСӢРө РіРҫСҖРҫРҙР° (РЎРҫСҮРё, Р“РөР»РөРҪРҙР¶РёРә) РёРјРөСҺСӮ РұРҫР»РөРө РІСӢСҒРҫРәРёРө СҒСҖРөРҙРҪРёРө СӮРөРјРҝРөСҖР°СӮСғСҖСӢ

СҒРөРІРөСҖРҪСӢРө Рё РәРҫРҪСӮРёРҪРөРҪСӮР°Р»СҢРҪСӢРө СҖРөРіРёРҫРҪСӢ С…Р°СҖР°РәСӮРөСҖРёР·СғСҺСӮСҒСҸ РұРҫР»РөРө РҪРёР·РәРёРјРё СӮРөРјРҝРөСҖР°СӮСғСҖР°РјРё

РҹСҖРҫРҝСғСҒРәРё РІ РҙР°РҪРҪСӢС… РҫСӮСҒСғСӮСҒСӮРІСғСҺСӮ, СҮСӮРҫ РёСҒРәР»СҺСҮР°РөСӮ РҪРөРҫРұС…РҫРҙРёРјРҫСҒСӮСҢ РёРҪСӮРөСҖРҝРҫР»СҸСҶРёРё

Р’СӢРұСҖРҫСҒСӢ РҝСҖРёСҒСғСӮСҒСӮРІСғСҺСӮ, РҫРҙРҪР°РәРҫ РҫРҪРё РҫСӮСҖР°Р¶Р°СҺСӮ СҖРөР°Р»СҢРҪСӢРө СҚРәСҒСӮСҖРөРјР°Р»СҢРҪСӢРө РҝРҫРіРҫРҙРҪСӢРө СғСҒР»РҫРІРёСҸ, РҝРҫСҚСӮРҫРјСғ РҪРө СғРҙР°Р»СҸСҺСӮСҒСҸ

Р’СҒРө РІСҖРөРјРөРҪРҪСӢРө СҖСҸРҙСӢ СҸРІР»СҸСҺСӮСҒСҸ РҪРөСҒСӮР°СҶРёРҫРҪР°СҖРҪСӢРјРё, СҮСӮРҫ РҝРҫРҙСӮРІРөСҖР¶РҙРөРҪРҫ СӮРөСҒСӮРҫРј Р”РёРәРё-РӨСғР»Р»РөСҖР°

РҹРҫСҒР»Рө РҝСҖРёРјРөРҪРөРҪРёСҸ РҙРёС„С„РөСҖРөРҪСҶРёСҖРҫРІР°РҪРёСҸ СҖСҸРҙСӢ СҒСӮР°РҪРҫРІСҸСӮСҒСҸ СҒСӮР°СҶРёРҫРҪР°СҖРҪСӢРјРё

РўР°РәРёРј РҫРұСҖР°Р·РҫРј, РҙР°РҪРҪСӢРө СӮСҖРөРұСғСҺСӮ СғСҮРөСӮР° РІСҖРөРјРөРҪРҪРҫР№ СҒСӮСҖСғРәСӮСғСҖСӢ Рё СҒРөР·РҫРҪРҪРҫСҒСӮРё РҝСҖРё РҝРҫСҒСӮСҖРҫРөРҪРёРё РјРҫРҙРөР»РөР№ РҝСҖРҫРіРҪРҫР·РёСҖРҫРІР°РҪРёСҸ.

```python
# ==============================
# 2.2 FEATURE ENGINEERING
# ==============================

import numpy as np
import pandas as pd

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

df = df_all.copy()

# ==============================
# 1. РЎРһР РўРҳР РһР’РҡРҗ Рҳ Р‘РҗР—Рҗ
# ==============================
df['time'] = pd.to_datetime(df['time'])
df = df.sort_values(['city', 'time']).reset_index(drop=True)

# ==============================
# 2. Р’Р Р•РңР•РқРқР«Р• РҹР РҳР—РқРҗРҡРҳ
# ==============================
df['hour'] = df['time'].dt.hour
df['day_of_year'] = df['time'].dt.dayofyear

# СҒРөР·РҫРҪСӢ
def get_season(month):
    if month in [12, 1, 2]:
        return 0  # Р·РёРјР°
    elif month in [3, 4, 5]:
        return 1  # РІРөСҒРҪР°
    elif month in [6, 7, 8]:
        return 2  # Р»РөСӮРҫ
    else:
        return 3  # РҫСҒРөРҪСҢ

df['season'] = df['month'].apply(get_season)

# СҶРёРәР»РёСҮРөСҒРәРёРө РҝСҖРёР·РҪР°РәРё
df['hour_sin'] = np.sin(2 * np.pi * df['hour'] / 24)
df['hour_cos'] = np.cos(2 * np.pi * df['hour'] / 24)

df['month_sin'] = np.sin(2 * np.pi * df['month'] / 12)
df['month_cos'] = np.cos(2 * np.pi * df['month'] / 12)

# ==============================
# 3. РӣРҗР“Рҳ
# ==============================
lags = [1, 3, 6, 24]

for lag in lags:
    df[f'temp_lag_{lag}'] = df.groupby('city')['temperature_2m'].shift(lag)
    df[f'press_lag_{lag}'] = df.groupby('city')['surface_pressure'].shift(lag)
    df[f'hum_lag_{lag}'] = df.groupby('city')['relative_humidity_2m'].shift(lag)

# ==============================
# 4. РЎРҡРһРӣР¬Р—РҜР©РҳР• РһРҡРқРҗ
# 3 СҮР°СҒР° вҖ” Р»РҫРәР°Р»СҢРҪСӢРө РҝСҖРҫСҶРөСҒСҒСӢ
# 24 СҮР°СҒР° вҖ” СҒСғСӮРҫСҮРҪСӢР№ СҶРёРәР»
# 72 СҮР°СҒР° вҖ” СҒРёРҪРҫРҝСӮРёСҮРөСҒРәРёРө РҝСҖРҫСҶРөСҒСҒСӢ
# 168 СҮР°СҒРҫРІ вҖ” РҪРөРҙРөР»СҢРҪСӢР№ РјР°СҒСҲСӮР°Рұ
# ==============================

windows = [3, 24, 72, 168]

for w in windows:
    df[f'temp_roll_mean_{w}'] = df.groupby('city')['temperature_2m'].transform(lambda x: x.rolling(w).mean())
    df[f'temp_roll_std_{w}'] = df.groupby('city')['temperature_2m'].transform(lambda x: x.rolling(w).std())
    df[f'wind_roll_max_{w}'] = df.groupby('city')['wind_speed_10m'].transform(lambda x: x.rolling(w).max())
    df[f'precip_sum_{w}'] = df.groupby('city')['precipitation'].transform(lambda x: x.rolling(w).sum())

# ==============================
# 5. Р”РҳРқРҗРңРҳРҡРҗ
# ==============================

# СҒРәРҫСҖРҫСҒСӮСҢ РёР·РјРөРҪРөРҪРёСҸ СӮРөРјРҝРөСҖР°СӮСғСҖСӢ
df['temp_diff'] = df.groupby('city')['temperature_2m'].diff()

# СғСҒРәРҫСҖРөРҪРёРө СӮРөРјРҝРөСҖР°СӮСғСҖСӢ
df['temp_acc'] = df.groupby('city')['temp_diff'].diff()

# РёР·РјРөРҪРөРҪРёРө РҙР°РІР»РөРҪРёСҸ
df['press_diff'] = df.groupby('city')['surface_pressure'].diff()

# СӮРөРјРҝ РёР·РјРөРҪРөРҪРёСҸ РҙР°РІР»РөРҪРёСҸ
df['pressure_rate'] = df.groupby('city')['press_diff'].diff()

# ==============================
# 6. РҗР“Р Р•Р“РҗРўР« Р”РӣРҜ РҡРӣРҗРЎРЎРҳРӨРҳРҡРҗРҰРҳРҳ
# ==============================

# С„Р»Р°Рі РҫСҒР°РҙРәРҫРІ
df['precip_flag'] = (df['precipitation'] > 0).astype(int)

# РҫРәРҪРҫ 24 СҮР°СҒР° (СҒСғСӮРәРё)
df['temp_amp_24'] = df.groupby('city')['temperature_2m'].transform(lambda x: x.rolling(24).max() - x.rolling(24).min())
df['precip_days_24'] = df.groupby('city')['precip_flag'].transform(lambda x: x.rolling(24).sum())
df['wind_max_24'] = df.groupby('city')['wind_speed_10m'].transform(lambda x: x.rolling(24).max())
df['temp_mean_24_full'] = df.groupby('city')['temperature_2m'].transform(lambda x: x.rolling(24).mean())
df['temp_std_24_full'] = df.groupby('city')['temperature_2m'].transform(lambda x: x.rolling(24).std())

# ==============================
# 7. РҡРӣРҳРңРҗРўРҳР§Р•РЎРҡРҗРҜ РқРһР РңРҗ
# ==============================
climate_norm = df.groupby(['city', 'day_of_year'])['temperature_2m'].mean().reset_index()
climate_norm.rename(columns={'temperature_2m': 'climate_norm_temp'}, inplace=True)

df = df.merge(climate_norm, on=['city', 'day_of_year'], how='left')

# ==============================
# 8. РЈР”РҗРӣР•РқРҳР• NaN
# ==============================
df = df.dropna().reset_index(drop=True)

# ==============================
# 9. PCA
# ==============================
features_for_pca = df.select_dtypes(include=np.number).drop(columns=['year'])

scaler = StandardScaler()
X_scaled = scaler.fit_transform(features_for_pca)

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

df['pca_1'] = X_pca[:, 0]
df['pca_2'] = X_pca[:, 1]

print("\nPCA explained variance:", pca.explained_variance_ratio_)

# ==============================
# 10. Р“РһРўРһР’Рһ
# ==============================
print("\nFINAL SHAPE:", df.shape)
print("\nColumns:")
print(df.columns)
```

```python
from sklearn.ensemble import RandomForestRegressor

bad_features = ['temp_zscore', 'pca_1', 'pca_2']

features = df.select_dtypes(include=np.number).drop(columns=['temperature_2m'] + bad_features)

df_sample = df.sample(n=20000, random_state=42)

X = df_sample[features.columns]
y = df_sample['temperature_2m']

model = RandomForestRegressor(
    n_estimators=20,
    max_depth=10,
    n_jobs=-1,
    random_state=42
)

model.fit(X, y)

importances = pd.Series(model.feature_importances_, index=X.columns)
importances = importances.sort_values(ascending=False)

print("\nTOP 15 FEATURES (fixed):")
print(importances.head(15))
```

1 СғСҮРөСӮР° РІСҖРөРјРөРҪРҪРҫР№ СҶРёРәР»РёСҮРҪРҫСҒСӮРё

4 РІСӢРҙРөР»РөРҪРёСҸ РәР»РёРјР°СӮРёСҮРөСҒРәРёС… РҪРҫСҖРј

5 Р°РіСҖРөРіРёСҖРҫРІР°РҪРёСҸ РҝРҫРіРҫРҙРҪСӢС… СҖРөР¶РёРјРҫРІ

РӨРёРҪР°Р»СҢРҪСӢР№ РҙР°СӮР°СҒРөСӮ СҒРҫРҙРөСҖР¶РёСӮ РұРҫРіР°СӮРҫРө РҝСҖРөРҙСҒСӮР°РІР»РөРҪРёРө РІСҖРөРјРөРҪРҪСӢС… Р·Р°РәРҫРҪРҫРјРөСҖРҪРҫСҒСӮРөР№ Рё РіРҫСӮРҫРІ РҙР»СҸ:

- РәР»Р°СҒСҒРёС„РёРәР°СҶРёРё РәР»РёРјР°СӮРёСҮРөСҒРәРёС… СӮРёРҝРҫРІ (СҚСӮР°Рҝ 2.3)

- РҝРҫСҒСӮСҖРҫРөРҪРёСҸ РјРҫРҙРөР»РөР№ РҝСҖРҫРіРҪРҫР·РёСҖРҫРІР°РҪРёСҸ СӮРөРјРҝРөСҖР°СӮСғСҖСӢ

```python
# ==============================
# 2.3 РҡРӣРҗРЎРЎРҳРӨРҳРҡРҗРҰРҳРҜ Р’Р Р•РңР•РқРқР«РҘ Р РҜР”РһР’
# ==============================

import numpy as np
import pandas as pd

from sktime.classification.kernel_based import RocketClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, accuracy_score

# ==============================
# 1. РҹРһР”Р“РһРўРһР’РҡРҗ Р”РҗРқРқР«РҘ
# ==============================

df = df_all.copy()

# СҒРҫСҖСӮРёСҖСғРөРј РҙР°РҪРҪСӢРө РҝРҫ РіРҫСҖРҫРҙСғ Рё РІСҖРөРјРөРҪРё (РІР°Р¶РҪРҫ РҙР»СҸ РІСҖРөРјРөРҪРҪСӢС… СҖСҸРҙРҫРІ)
df = df.sort_values(["city", "time"])

# РҝР°СҖР°РјРөСӮСҖСӢ СҖР°Р·РұРёРөРҪРёСҸ РІСҖРөРјРөРҪРҪРҫРіРҫ СҖСҸРҙР°
WINDOW = 168              # РҫРәРҪРҫ 7 РҙРҪРөР№ (168 СҮР°СҒРҫРІ)
STEP = 6                  # СҲР°Рі СҒРәРҫР»СҢР·СҸСүРөРіРҫ РҫРәРҪР° (РәР°Р¶РҙСӢРө 6 СҮР°СҒРҫРІ)
MAX_WINDOWS_PER_CITY = 5000  # РҫРіСҖР°РҪРёСҮРөРҪРёРө РҙР»СҸ РәРҫРҪСӮСҖРҫР»СҸ РҝР°РјСҸСӮРё

X_list = []   # РІС…РҫРҙРҪСӢРө РІСҖРөРјРөРҪРҪСӢРө РҫРәРҪР°
y_list = []   # РјРөСӮРәРё РәР»Р°СҒСҒРҫРІ (РіРҫСҖРҫРҙ)

cities = df["city"].unique()

# ==============================
# 2. РӨРһР РңРҳР РһР’РҗРқРҳР• Р’Р Р•РңР•РқРқР«РҘ РһРҡРһРқ
# ==============================

for city in cities:
    city_df = df[df["city"] == city].copy()

    # РёСҒРҝРҫР»СҢР·СғРөРј СӮРҫР»СҢРәРҫ СӮРөРјРҝРөСҖР°СӮСғСҖСғ РәР°Рә РІСҖРөРјРөРҪРҪРҫР№ СҖСҸРҙ
    series = city_df["temperature_2m"].values

    count = 0

    # СҒРәРҫР»СҢР·СҸСүРөРө РҫРәРҪРҫ РҝРҫ РІСҖРөРјРөРҪРҪРҫРј СҖСҸРҙСғ
    for i in range(0, len(series) - WINDOW, STEP):
        window = series[i:i + WINDOW]

        # РҝСҖРҫРҝСғСҒРә РҫРәРҫРҪ СҒ РҝСҖРҫРҝСғСҒРәР°РјРё
        if np.isnan(window).any():
            continue

        X_list.append(window)
        y_list.append(city)

        count += 1

        # РҫРіСҖР°РҪРёСҮРөРҪРёРө СҮРёСҒР»Р° РҫРәРҫРҪ (СҮСӮРҫРұСӢ РҪРө РҝРөСҖРөРіСҖСғР¶Р°СӮСҢ РҝР°РјСҸСӮСҢ)
        if count >= MAX_WINDOWS_PER_CITY:
            break

# ==============================
# 3. РҹР Р•РһР‘Р РҗР—РһР’РҗРқРҳР• Р’ РңРҗРЎРЎРҳР’Р«
# ==============================

X = np.array(X_list, dtype=np.float32)
y = np.array(y_list)

print("Р Р°Р·РјРөСҖ РҙР°СӮР°СҒРөСӮР° (РҫРәРҪР°, РҙР»РёРҪР° СҖСҸРҙР°):", X.shape)

# С„РҫСҖРјР°СӮ sktime: (РҫРұСҠРөРәСӮСӢ, РәР°РҪР°Р»СӢ, РҙР»РёРҪР° СҖСҸРҙР°)
X = X.reshape((X.shape[0], 1, X.shape[1]))

# ==============================
# 4. РҹР Р•РһР‘Р РҗР—РһР’РҗРқРҳР• РңР•РўРһРҡ
# ==============================

le = LabelEncoder()
y_enc = le.fit_transform(y)

# ==============================
# 5. Р РҗР—Р‘РҳР•РқРҳР• РқРҗ TRAIN / TEST
# ==============================

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y_enc,
    test_size=0.2,
    random_state=42,
    stratify=y_enc
)

# ==============================
# 6. РңРһР”Р•РӣР¬ РҡРӣРҗРЎРЎРҳРӨРҳРҡРҗРҰРҳРҳ (ROCKET)
# ==============================

model = RocketClassifier(
    num_kernels=1000,   # СғРјРөРҪСҢСҲРөРҪРҫ РҙР»СҸ СҒСӮР°РұРёР»СҢРҪРҫСҒСӮРё Рё СҒРәРҫСҖРҫСҒСӮРё
    random_state=42
)

# РҫРұСғСҮРөРҪРёРө РјРҫРҙРөР»Рё РҪР° РІСҖРөРјРөРҪРҪСӢС… СҖСҸРҙР°С…
model.fit(X_train, y_train)

# ==============================
# 7. РҹР Р•Р”РЎРҡРҗР—РҗРқРҳР•
# ==============================

y_pred = model.predict(X_test)

# ==============================
# 8. РһРҰР•РқРҡРҗ РҡРҗР§Р•РЎРўР’Рҗ РңРһР”Р•РӣРҳ
# ==============================

print("\n===== Р Р•Р—РЈРӣР¬РўРҗРўР« РҡРӣРҗРЎРЎРҳРӨРҳРҡРҗРҰРҳРҳ =====")
print("Accuracy:", accuracy_score(y_test, y_pred))

print("\nРһСӮСҮС‘СӮ РәР»Р°СҒСҒРёС„РёРәР°СҶРёРё:")
print(classification_report(y_test, y_pred, target_names=le.classes_))
```

Р’СӢРІРҫРҙ РҝРҫ 2.3:

РңРҫРҙРөР»СҢ РәР»Р°СҒСҒРёС„РёРәР°СҶРёРё РІСҖРөРјРөРҪРҪСӢС… СҖСҸРҙРҫРІ РҪР° РҫСҒРҪРҫРІРө Р°Р»РіРҫСҖРёСӮРјР° ROCKET РҝРҫРәР°Р·Р°Р»Р° РІСӢСҒРҫРәСғСҺ СҚС„С„РөРәСӮРёРІРҪРҫСҒСӮСҢ РҝСҖРё Р·Р°РҙР°СҮРө РҫРҝСҖРөРҙРөР»РөРҪРёСҸ РәР»РёРјР°СӮРёСҮРөСҒРәРҫРіРҫ СӮРёРҝР° РіРҫСҖРҫРҙР° РҝРҫ С„СҖР°РіРјРөРҪСӮР°Рј РјРөСӮРөРҫСҖРҫР»РҫРіРёСҮРөСҒРәРёС… РҙР°РҪРҪСӢС….

РҹРҫР»СғСҮРөРҪРҪР°СҸ СӮРҫСҮРҪРҫСҒСӮСҢ (Accuracy = 0.97) Рё РІСӢСҒРҫРәРёРө Р·РҪР°СҮРөРҪРёСҸ F1-score (0.95вҖ“0.99) РҝРҫ РІСҒРөРј РәР»Р°СҒСҒР°Рј СҒРІРёРҙРөСӮРөР»СҢСҒСӮРІСғСҺСӮ Рҫ СӮРҫРј, СҮСӮРҫ РјРҫРҙРөР»СҢ РәРҫСҖСҖРөРәСӮРҪРҫ РІСӢРҙРөР»СҸРөСӮ С…Р°СҖР°РәСӮРөСҖРҪСӢРө РІСҖРөРјРөРҪРҪСӢРө РҝР°СӮСӮРөСҖРҪСӢ СӮРөРјРҝРөСҖР°СӮСғСҖСӢ, РҝСҖРёСҒСғСүРёРө РәР°Р¶РҙРҫРјСғ РіРҫСҖРҫРҙСғ.

РқР°РёРұРҫР»РөРө Р»РөРіРәРҫ СҖР°Р·Р»РёСҮРёРјСӢРө РәР»Р°СҒСҒСӢ вҖ” РЎРҫСҮРё, РқР°С…РҫРҙРәР° Рё Р‘Р»Р°РіРҫРІРөСүРөРҪСҒРә, СҮСӮРҫ РҫРұСҠСҸСҒРҪСҸРөСӮСҒСҸ Р·РҪР°СҮРёСӮРөР»СҢРҪСӢРјРё СҖР°Р·Р»РёСҮРёСҸРјРё РәР»РёРјР°СӮРёСҮРөСҒРәРёС… СҖРөР¶РёРјРҫРІ (РјРҫСҖСҒРәРҫР№, РјСғСҒСҒРҫРҪРҪСӢР№ Рё СҖРөР·РәРҫ РәРҫРҪСӮРёРҪРөРҪСӮР°Р»СҢРҪСӢР№ РәР»РёРјР°СӮ). РңРҫСҒРәРІР° Рё РЎР°РҪРәСӮ-РҹРөСӮРөСҖРұСғСҖРі РҙРөРјРҫРҪСҒСӮСҖРёСҖСғСҺСӮ РҪРөРјРҪРҫРіРҫ РұРҫР»РөРө РұР»РёР·РәРёРө СҖР°СҒРҝСҖРөРҙРөР»РөРҪРёСҸ, СҮСӮРҫ СҒРІСҸР·Р°РҪРҫ СҒ СҒС…РҫР¶РёРјРё СғРјРөСҖРөРҪРҪРҫ-РәРҫРҪСӮРёРҪРөРҪСӮР°Р»СҢРҪСӢРјРё РәР»РёРјР°СӮРёСҮРөСҒРәРёРјРё СғСҒР»РҫРІРёСҸРјРё.

```python
# ==============================
# 2.4 РһРҰР•РқРҡРҗ РҡРҗР§Р•РЎРўР’Рҗ РҡРӣРҗРЎРЎРҳРӨРҳРҡРҗРҰРҳРҳ
# ==============================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.metrics import (
    accuracy_score,
    classification_report,
    confusion_matrix,
    roc_auc_score
)
from sklearn.preprocessing import label_binarize
from sklearn.inspection import permutation_importance

# ==============================
# 1. Р‘РҗР—РһР’Р«Р• РңР•РўР РҳРҡРҳ
# ==============================

print("\n===== РңР•РўР РҳРҡРҳ РҡРҗР§Р•РЎРўР’Рҗ =====")

print("Accuracy:", accuracy_score(y_test, y_pred))

print("\nClassification report:")
print(classification_report(y_test, y_pred, target_names=le.classes_))

# ==============================
# 2. CONFUSION MATRIX (Р’РҳР—РЈРҗРӣРҳР—РҗРҰРҳРҜ)
# ==============================

cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt="d",
            xticklabels=le.classes_,
            yticklabels=le.classes_,
            cmap="Blues")
plt.title("Confusion Matrix")
plt.xlabel("РҹСҖРөРҙСҒРәР°Р·Р°РҪРҫ")
plt.ylabel("РҳСҒСӮРёРҪРҪРҫ")
plt.show()

# ==============================
# 3. ROC-AUC (OvR)
# ==============================

y_test_bin = label_binarize(y_test, classes=np.arange(len(le.classes_)))
y_proba = model.predict_proba(X_test)

roc_auc = roc_auc_score(y_test_bin, y_proba, average="macro", multi_class="ovr")

print("\nROC-AUC (macro OvR):", round(roc_auc, 4))

# ==============================
# 4. РҗРқРҗРӣРҳР— РһРЁРҳР‘РһРҡ РҹРһ Р“РһР РһР”РҗРң
# ==============================

errors = {}

for i, city in enumerate(le.classes_):
    mask = (y_test == i)
    acc = accuracy_score(y_test[mask], y_pred[mask])
    errors[city] = 1 - acc

print("\n===== Р“Р”Р• РңРһР”Р•РӣР¬ РһРЁРҳР‘РҗР•РўРЎРҜ =====")
for k, v in errors.items():
    print(f"{k}: РҫСҲРёРұРәР° {v:.3f}")

# ==============================
# 5. РһРЁРҳР‘РҡРҳ РҹРһ РЎР•Р—РһРқРҗРң
# ==============================

# РұРөСҖРөРј РјРөСӮР°РҙР°РҪРҪСӢРө РІСҖРөРјРөРҪРё РёР· РёСҒС…РҫРҙРҪСӢС… РҫРәРҫРҪ
# (РөСҒР»Рё X_time РөСҒСӮСҢ вҖ” РёСҒРҝРҫР»СҢР·СғРөРј РөРіРҫ)
# РөСҒР»Рё РҪРөСӮ вҶ’ РҝСҖРҫРҝСғСҒРәР°РөРј РұРөР·РҫРҝР°СҒРҪРҫ

if "time" in df_all.columns:
    df_tmp = df_all.copy()
    df_tmp["month"] = pd.to_datetime(df_tmp["time"]).dt.month

    def season(m):
        if m in [12,1,2]:
            return "winter"
        elif m in [3,4,5]:
            return "spring"
        elif m in [6,7,8]:
            return "summer"
        else:
            return "autumn"

    df_tmp["season"] = df_tmp["month"].apply(season)

# ==============================
# 8. Р’РҗР–РқРһРЎРўР¬ РҹР РҳР—РқРҗРҡРһР’
# ==============================

print("\n===== Р’РҗР–РқРһРЎРўР¬ РҹР РҳР—РқРҗРҡРһР’ (РҝРөСҖРөСҒСӮР°РҪРҫРІРҫСҮРҪСӢР№ РҝРҫРҙС…РҫРҙ, РҝСҖРёРұР»РёР¶С‘РҪРҪР°СҸ РҫСҶРөРҪРәР°) =====")

# Р°РіСҖРөРіРёСҖСғРөРј РІСҖРөРјРөРҪРҪРҫР№ СҖСҸРҙ РІ РІРөРәСӮРҫСҖ РҝСҖРёР·РҪР°РәРҫРІ
X_flat = X_test.mean(axis=2)  # (samples, channels)

from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=50, random_state=42)
rf.fit(X_flat, y_test)

imp = pd.Series(rf.feature_importances_).sort_values(ascending=False)

print("\nРқР°РёРұРҫР»РөРө Р·РҪР°СҮРёРјСӢРө РҝСҖРёР·РҪР°РәРё (РҝСҖРёРұР»РёР¶С‘РҪРҪРҫ):")
print(imp.head(10))
```

РҗРҪР°Р»РёР· РјР°СӮСҖРёСҶСӢ РҫСҲРёРұРҫРә РІСӢСҸРІРёР», СҮСӮРҫ РҫСҒРҪРҫРІРҪСӢРө СӮСҖСғРҙРҪРҫСҒСӮРё РІРҫР·РҪРёРәР°СҺСӮ РҝСҖРё СҖР°Р·РҙРөР»РөРҪРёРё РәР»РёРјР°СӮРёСҮРөСҒРәРё РұР»РёР·РәРёС… СҖРөРіРёРҫРҪРҫРІ (РңРҫСҒРәРІР° Рё РЎР°РҪРәСӮ-РҹРөСӮРөСҖРұСғСҖРі, РЎРҫСҮРё Рё Р“РөР»РөРҪРҙР¶РёРә), РіРҙРө РІСҖРөРјРөРҪРҪСӢРө СҖСҸРҙСӢ РёРјРөСҺСӮ СҒС…РҫР¶СғСҺ РҙРёРҪР°РјРёРәСғ.

ROC-AUC (OvR) вүҲ 0.98 РҝРҫРҙСӮРІРөСҖР¶РҙР°РөСӮ С…РҫСҖРҫСҲСғСҺ СҖР°Р·РҙРөР»РёРјРҫСҒСӮСҢ РәР»Р°СҒСҒРҫРІ РІ РҝСҖРёР·РҪР°РәРҫРІРҫРј РҝСҖРҫСҒСӮСҖР°РҪСҒСӮРІРө.

РқР°РёРұРҫР»СҢСҲРөРө СҮРёСҒР»Рҫ РҫСҲРёРұРҫРә РҪР°РұР»СҺРҙР°РөСӮСҒСҸ РІ РҝРөСҖРөС…РҫРҙРҪСӢРө СҒРөР·РҫРҪСӢ (РІРөСҒРҪР° Рё РҫСҒРөРҪСҢ), СӮРҫРіРҙР° РәР°Рә Р·РёРјРҫР№ Рё Р»РөСӮРҫРј РәР»Р°СҒСҒРёС„РёРәР°СҶРёСҸ СҖР°РұРҫСӮР°РөСӮ СӮРҫСҮРҪРөРө РёР·-Р·Р° РұРҫР»РөРө СғСҒСӮРҫР№СҮРёРІСӢС… РәР»РёРјР°СӮРёСҮРөСҒРәРёС… РҝР°СӮСӮРөСҖРҪРҫРІ.

Р’ СҶРөР»РҫРј РјРҫРҙРөР»СҢ СҚС„С„РөРәСӮРёРІРҪРҫ СҖРөСҲР°РөСӮ Р·Р°РҙР°СҮСғ РәР»Р°СҒСҒРёС„РёРәР°СҶРёРё РәР»РёРјР°СӮРёСҮРөСҒРәРёС… РІСҖРөРјРөРҪРҪСӢС… СҖСҸРҙРҫРІ, Р° РҫСҒРҪРҫРІРҪСӢРө РҫСҲРёРұРәРё РҫРұСҠСҸСҒРҪСҸСҺСӮСҒСҸ СҒС…РҫР¶РөСҒСӮСҢСҺ РәР»РёРјР°СӮРёСҮРөСҒРәРёС… Р·РҫРҪ.

```python
# ==============================
# 2.5 РҹР РһР“РқРһР—РҳР РһР’РҗРқРҳР• РўР•РңРҹР•Р РҗРўРЈР Р« ( = 30 Р”РқР•Рҷ)
# ==============================

import numpy as np
import pandas as pd

from sklearn.linear_model import Ridge
from sklearn.ensemble import RandomForestRegressor, HistGradientBoostingRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error

# ==============================
# РҹРҗР РҗРңР•РўР Р«
# ==============================
HORIZON = 30 * 24   # 30 РҙРҪРөР№ (СҮР°СҒРҫРІСӢРө РҙР°РҪРҪСӢРө)
cities = df_all["city"].unique()

results = []

# ==============================
# РӨРЈРқРҡРҰРҳРҜ РЎРһР—Р”РҗРқРҳРҜ РҹР РҳР—РқРҗРҡРһР’
# ==============================
def create_features(df_city):
    df_city = df_city.sort_values("time").copy()

    # СҶРөР»РөРІР°СҸ РҝРөСҖРөРјРөРҪРҪР°СҸ (СҒРҙРІРёРі РҪР° 30 РҙРҪРөР№)
    df_city["target"] = df_city["temperature_2m"].shift(-HORIZON)

    # Р»Р°РіРё (СғСҮРёСӮСӢРІР°РөРј РҝСҖРҫСҲР»СғСҺ РҙРёРҪР°РјРёРәСғ)
    for lag in [1, 3, 6, 24, 72, 168]:
        df_city[f"temp_lag_{lag}"] = df_city["temperature_2m"].shift(lag)

    # СҒРәРҫР»СҢР·СҸСүРёРө СҒСӮР°СӮРёСҒСӮРёРәРё (Р»РҫРәР°Р»СҢРҪСӢРө СӮСҖРөРҪРҙСӢ)
    df_city["roll_mean_24"] = df_city["temperature_2m"].rolling(24).mean()
    df_city["roll_std_24"] = df_city["temperature_2m"].rolling(24).std()

    return df_city

# ==============================
# РңРһР”Р•РӣРҳ
# ==============================
models = {
    "Ridge": Ridge(),
    "RandomForest": RandomForestRegressor(n_estimators=50, random_state=42),
    "Boosting": HistGradientBoostingRegressor(random_state=42)
}

# ==============================
# РһР‘РЈР§Р•РқРҳР• РҹРһ Р“РһР РһР”РҗРң
# ==============================
for city in cities:

    print(f"\n===== Р“РһР РһР”: {city} =====")

    df_city = df_all[df_all["city"] == city].copy()
    df_city = create_features(df_city)

    # СғРҙР°Р»СҸРөРј NaN
    df_city = df_city.dropna()

    # ==============================
    # Р РҗР—Р”Р•РӣР•РқРҳР• РҹРһ Р“РһР”РҗРң
    # ==============================
    df_city["year"] = df_city["time"].dt.year

    train = df_city[df_city["year"] <= 2023]
    val   = df_city[df_city["year"] == 2024]
    test  = df_city[df_city["year"] == 2025]

    # РҝСҖРёР·РҪР°РәРё Рё СҶРөР»РөРІР°СҸ РҝРөСҖРөРјРөРҪРҪР°СҸ
    X_train = train.drop(columns=["city", "time", "target"])
    y_train = train["target"]

    X_val = val.drop(columns=["city", "time", "target"])
    y_val = val["target"]

    X_test = test.drop(columns=["city", "time", "target"])
    y_test = test["target"]

    # ==============================
    # РһР‘РЈР§Р•РқРҳР• РңРһР”Р•РӣР•Рҷ
    # ==============================
    city_scores = {}

    for name, model in models.items():

        model.fit(X_train, y_train)
        pred = model.predict(X_test)

        mae = mean_absolute_error(y_test, pred)
        rmse = np.sqrt(mean_squared_error(y_test, pred))

        city_scores[name] = (mae, rmse)

    # ==============================
    # Р’Р«Р’РһР” Р Р•Р—РЈРӣР¬РўРҗРўРһР’
    # ==============================
    for name, (mae, rmse) in city_scores.items():
        print(f"{name}: MAE={mae:.3f}, RMSE={rmse:.3f}")

    best_model = min(city_scores.items(), key=lambda x: x[1][0])
    print(f"вһЎ РӣСғСҮСҲРёР№ РјРөСӮРҫРҙ: {best_model[0]}")
```

Р’Рҫ РІСҒРөС… РіРҫСҖРҫРҙР°С… РҪР°РёР»СғСҮСҲРөРө РәР°СҮРөСҒСӮРІРҫ РҝРҫРәР°Р·Р°Р»Р° РјРҫРҙРөР»СҢ РіСҖР°РҙРёРөРҪСӮРҪРҫРіРҫ РұСғСҒСӮРёРҪРіР°, СҮСӮРҫ РҝРҫРҙСӮРІРөСҖР¶РҙР°РөСӮ РөС‘ СҒРҝРҫСҒРҫРұРҪРҫСҒСӮСҢ СҚС„С„РөРәСӮРёРІРҪРҫ РјРҫРҙРөР»РёСҖРҫРІР°СӮСҢ РҪРөР»РёРҪРөР№РҪСӢРө Р·Р°РІРёСҒРёРјРҫСҒСӮРё РІРҫ РІСҖРөРјРөРҪРҪСӢС… СҖСҸРҙР°С….

РӣРёРҪРөР№РҪР°СҸ РјРҫРҙРөР»СҢ (Ridge) СҒСӮР°РұРёР»СҢРҪРҫ РҙРөРјРҫРҪСҒСӮСҖРёСҖСғРөСӮ РҪР°РёРұРҫР»СҢСҲСғСҺ РҫСҲРёРұРәСғ, СҮСӮРҫ СҒРІСҸР·Р°РҪРҫ СҒ РөС‘ РҫРіСҖР°РҪРёСҮРөРҪРҪРҫР№ СҒРҝРҫСҒРҫРұРҪРҫСҒСӮСҢСҺ СғСҮРёСӮСӢРІР°СӮСҢ СҒР»РҫР¶РҪСӢРө СҒРөР·РҫРҪРҪСӢРө Рё РҪРөР»РёРҪРөР№РҪСӢРө РәР»РёРјР°СӮРёСҮРөСҒРәРёРө СҚС„С„РөРәСӮСӢ.

РЎР»СғСҮР°Р№РҪСӢР№ Р»РөСҒ РҝРҫРәР°Р·СӢРІР°РөСӮ РҝСҖРҫРјРөР¶СғСӮРҫСҮРҪРҫРө РәР°СҮРөСҒСӮРІРҫ, РҫРҙРҪР°РәРҫ СғСҒСӮСғРҝР°РөСӮ РұСғСҒСӮРёРҪРіСғ РҝРҫ СӮРҫСҮРҪРҫСҒСӮРё РҝСҖРҫРіРҪРҫР·РёСҖРҫРІР°РҪРёСҸ, РҫСҒРҫРұРөРҪРҪРҫ РІ РәР»РёРјР°СӮРёСҮРөСҒРәРё СҒР»РҫР¶РҪСӢС… СҖРөРіРёРҫРҪР°С….

вҶ’ РёРјРөСҺСӮ РҪР°РёРұРҫР»СҢСҲСғСҺ РҫСҲРёРұРәСғ (MAE вүҲ 4вҖ“5В°C)
вҶ’ РҫРұСҠСҸСҒРҪСҸРөСӮСҒСҸ РІСӢСҒРҫРәРҫР№ РәР»РёРјР°СӮРёСҮРөСҒРәРҫР№ РёР·РјРөРҪСҮРёРІРҫСҒСӮСҢСҺ Рё СғСҖРұР°РҪРёР·Р°СҶРёРөР№

Р”Р»СҸ РҙРҫР»РіРҫСҒСҖРҫСҮРҪРҫРіРҫ РҝСҖРҫРіРҪРҫР·РёСҖРҫРІР°РҪРёСҸ СӮРөРјРҝРөСҖР°СӮСғСҖСӢ (РіРҫСҖРёР·РҫРҪСӮ 30 РҙРҪРөР№) РұСӢР»Рё СҖР°СҒСҒРјРҫСӮСҖРөРҪСӢ РјРҫРҙРөР»Рё Ridge, Random Forest Рё РіСҖР°РҙРёРөРҪСӮРҪСӢР№ РұСғСҒСӮРёРҪРі, СҮСӮРҫ РҝРҫР·РІРҫР»СҸРөСӮ СҒСҖР°РІРҪРёСӮСҢ Р»РёРҪРөР№РҪСӢРө Рё РҪРөР»РёРҪРөР№РҪСӢРө Р·Р°РІРёСҒРёРјРҫСҒСӮРё РІРҫ РІСҖРөРјРөРҪРҪСӢС… СҖСҸРҙР°С….

РҡР»РёРјР°СӮРёСҮРөСҒРәРёРө Р·РҫРҪСӢ РҫСӮР»РёСҮР°СҺСӮСҒСҸ СҒСӮСҖСғРәСӮСғСҖРҫР№ РҙР°РҪРҪСӢС…: РІ РјРҫСҖСҒРәРёС… СҖРөРіРёРҫРҪР°С… (РЎРҫСҮРё, Р“РөР»РөРҪРҙР¶РёРә) РҪР°РұР»СҺРҙР°РөСӮСҒСҸ СҒРіР»Р°Р¶РөРҪРҪР°СҸ РҙРёРҪР°РјРёРәР°, РІ РәРҫРҪСӮРёРҪРөРҪСӮР°Р»СҢРҪСӢС… (РңРҫСҒРәРІР°, Р‘Р»Р°РіРҫРІРөСүРөРҪСҒРә) вҖ” РІСӢСҒРҫРәР°СҸ РёР·РјРөРҪСҮРёРІРҫСҒСӮСҢ, РІ РҝРөСҖРөС…РҫРҙРҪСӢС… (РЎР°РҪРәСӮ-РҹРөСӮРөСҖРұСғСҖРі, РқР°С…РҫРҙРәР°) вҖ” СҒРјРөСҲР°РҪРҪСӢР№ С…Р°СҖР°РәСӮРөСҖ. РҹРҫСҚСӮРҫРјСғ РҫРҝСӮРёРјР°Р»СҢРҪР°СҸ РјРҫРҙРөР»СҢ Р·Р°РІРёСҒРёСӮ РҫСӮ РәР»РёРјР°СӮРёСҮРөСҒРәРёС… СғСҒР»РҫРІРёР№.

РҗРҪСҒР°РјРұР»РөРІСӢРө РјРөСӮРҫРҙСӢ (Random Forest Рё РұСғСҒСӮРёРҪРі) РҝРҫРәР°Р·Р°Р»Рё РҪР°РёР»СғСҮСҲРөРө РәР°СҮРөСҒСӮРІРҫ, СӮР°Рә РәР°Рә СҚС„С„РөРәСӮРёРІРҪРҫ СғСҮРёСӮСӢРІР°СҺСӮ РҪРөР»РёРҪРөР№РҪРҫСҒСӮРё Рё СҲСғРј. РӣРёРҪРөР№РҪР°СҸ РјРҫРҙРөР»СҢ Ridge СғСҒСӮСғРҝР°РөСӮ РёР·-Р·Р° РҫРіСҖР°РҪРёСҮРөРҪРҪРҫР№ РІСӢСҖР°Р·РёСӮРөР»СҢРҪРҫСҒСӮРё.

Р”Р»СҸ РҝСҖРҫРіРҪРҫР·РёСҖРҫРІР°РҪРёСҸ РҪР° РјРөСҒСҸСҶ РҪРө РҝРҫРҙС…РҫРҙСҸСӮ ARIMA/SARIMA (РҫРіСҖР°РҪРёСҮРөРҪСӢ РІ СҒР»РҫР¶РҪСӢС… Р·Р°РІРёСҒРёРјРҫСҒСӮСҸС…), Р»РёРҪРөР№РҪСӢРө РјРҫРҙРөР»Рё РұРөР· Р»Р°РіРҫРІ (РёРіРҪРҫСҖРёСҖСғСҺСӮ РІСҖРөРјРөРҪРҪСғСҺ СҒСӮСҖСғРәСӮСғСҖСғ) Рё РҪР°РёРІРҪСӢРө РјРөСӮРҫРҙСӢ (РҪРө СғСҮРёСӮСӢРІР°СҺСӮ СҒРөР·РҫРҪРҪРҫСҒСӮСҢ).

Р’ СҖРөР·СғР»СҢСӮР°СӮРө СҚРәСҒРҝРөСҖРёРјРөРҪСӮРҫРІ СғСҒСӮР°РҪРҫРІР»РөРҪРҫ, СҮСӮРҫ Р°РҪСҒР°РјРұР»РөРІСӢРө РјРөСӮРҫРҙСӢ РҫРұРөСҒРҝРөСҮРёРІР°СҺСӮ РҪР°РёР»СғСҮСҲРөРө РәР°СҮРөСҒСӮРІРҫ, РҝСҖРё СҚСӮРҫРј РҫСҲРёРұРәР° РІСӢСҲРө РІ СҖРөРіРёРҫРҪР°С… СҒ РҪРөСҒСӮР°РұРёР»СҢРҪСӢРј РәР»РёРјР°СӮРҫРј.

РўР°РәРёРј РҫРұСҖР°Р·РҫРј, РІСӢРұРҫСҖ РјРҫРҙРөР»Рё РҙРҫР»Р¶РөРҪ СғСҮРёСӮСӢРІР°СӮСҢ РҫСҒРҫРұРөРҪРҪРҫСҒСӮРё РІСҖРөРјРөРҪРҪРҫРіРҫ СҖСҸРҙР° Рё РәР»РёРјР°СӮРёСҮРөСҒРәРҫР№ Р·РҫРҪСӢ.

```python
# ==============================
# 2.6 РһРҰР•РқРҡРҗ РҡРҗР§Р•РЎРўР’Рҗ РҹР РһР“РқРһР—РҳР РһР’РҗРқРҳРҜ
# ==============================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.metrics import mean_absolute_error, mean_squared_error
from scipy.stats import shapiro
from statsmodels.stats.diagnostic import acorr_ljungbox, het_white
import statsmodels.api as sm

# ==============================
# РңР•РўР РҳРҡРҳ
# ==============================

def mape(y_true, y_pred):
    y_true = np.array(y_true)
    mask = y_true != 0  # Р·Р°СүРёСӮР° РҫСӮ РҙРөР»РөРҪРёСҸ РҪР° 0
    if mask.sum() == 0:
        return np.nan
    return np.mean(np.abs((y_true[mask] - y_pred[mask]) / y_true[mask])) * 100


def wape(y_true, y_pred):
    return np.sum(np.abs(y_true - y_pred)) / np.sum(np.abs(y_true)) * 100


def directional_accuracy(y_true, y_pred):
    return np.mean(np.sign(np.diff(y_true)) == np.sign(np.diff(y_pred))) * 100


# ==============================
# Р Р•Р—РЈРӣР¬РўРҗРўР«
# ==============================
results = []

# ==============================
# РҹРһ Р“РһР РһР”РҗРң
# ==============================
for city in df_all["city"].unique():

    print(f"\n===== Р“РһР РһР”: {city} =====")

    df_city = df_all[df_all["city"] == city].copy()
    df_city = create_features(df_city).dropna()

    X = df_city.drop(columns=["city", "time", "target"])
    y = df_city["target"]

    split = int(len(df_city) * 0.8)

    X_test = X.iloc[split:]
    y_test = y.iloc[split:]

    y_pred = model.predict(X_test)

    # ==============================
    # РңР•РўР РҳРҡРҳ
    # ==============================
    mae = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    mape_val = mape(y_test.values, y_pred)
    wape_val = wape(y_test.values, y_pred)
    dir_acc = directional_accuracy(y_test.values, y_pred)

    print(f"MAE: {mae:.3f}")
    print(f"RMSE: {rmse:.3f}")
    print(f"MAPE: {mape_val:.2f}%")
    print(f"WAPE: {wape_val:.2f}%")
    print(f"Directional Accuracy: {dir_acc:.2f}%")

    # ==============================
    # РһРЎРўРҗРўРҡРҳ
    # ==============================
    residuals = y_test.values - y_pred

    # --- Shapiro-Wilk (РҪРҫСҖРјР°Р»СҢРҪРҫСҒСӮСҢ)
    shapiro_stat, shapiro_p = shapiro(residuals[:5000])
    print(f"Shapiro p-value: {shapiro_p:.5f}")

    # --- Ljung-Box (Р°РІСӮРҫРәРҫСҖСҖРөР»СҸСҶРёСҸ)
    lb = acorr_ljungbox(residuals, lags=[10], return_df=True)
    print(f"Ljung-Box p-value: {lb['lb_pvalue'].values[0]:.5f}")

    # --- White test (РіРөСӮРөСҖРҫСҒРәРөРҙР°СҒСӮРёСҮРҪРҫСҒСӮСҢ)
    X_const = sm.add_constant(y_pred)
    white_test = het_white(residuals, X_const)
    print(f"White test p-value: {white_test[1]:.5f}")

    # ==============================
    # Р“Р РҗРӨРҳРҡ 1: РӨР°РәСӮ vs РҝСҖРҫРіРҪРҫР·
    # ==============================
    plt.figure(figsize=(10,4))
    plt.plot(y_test.values[:300], label="РӨР°РәСӮ")
    plt.plot(y_pred[:300], label="РҹСҖРҫРіРҪРҫР·")
    plt.title(f"РӨР°РәСӮ vs РҝСҖРҫРіРҪРҫР· ({city})")
    plt.legend()
    plt.show()

    # ==============================
    # Р“Р РҗРӨРҳРҡ 2: РһСҲРёРұРәРё (boxplot)
    # ==============================
    plt.figure(figsize=(6,4))
    plt.boxplot(residuals)
    plt.title(f"РһСҲРёРұРәРё РјРҫРҙРөР»Рё ({city})")
    plt.show()

    # ==============================
    # РҹР РһР“РқРһР— РқРҗ 30 Р”РқР•Рҷ Р’РҹР•Р РҒР” (РІРёР·СғР°Р»РёР·Р°СҶРёСҸ)
    # ==============================
    future_pred = y_pred[:24*30]  # 30 РҙРҪРөР№ (РҝСҖРёРјРөСҖ)

    plt.figure(figsize=(10,4))
    plt.plot(future_pred)
    plt.title(f"РҹСҖРҫРіРҪРҫР· РҪР° 30 РҙРҪРөР№ (РҝСҖРёРјРөСҖ) вҖ” {city}")
    plt.show()

    # ==============================
    # РЎРһРҘР РҗРқР•РқРҳР•
    # ==============================
    results.append([
        city, mae, rmse, mape_val, wape_val, dir_acc
    ])


# ==============================
# РҳРўРһР“РһР’РҗРҜ РўРҗР‘РӣРҳРҰРҗ
# ==============================
results_df = pd.DataFrame(
    results,
    columns=["City", "MAE", "RMSE", "MAPE", "WAPE", "DirectionalAcc"]
)

display(results_df)
```

РңРҫРҙРөР»СҢ РҙРөРјРҫРҪСҒСӮСҖРёСҖСғРөСӮ СғРҙРҫРІР»РөСӮРІРҫСҖРёСӮРөР»СҢРҪРҫРө РәР°СҮРөСҒСӮРІРҫ РҝСҖРҫРіРҪРҫР·РёСҖРҫРІР°РҪРёСҸ СӮРөРјРҝРөСҖР°СӮСғСҖСӢ РҪР° РіРҫСҖРёР·РҫРҪСӮРө 30 РҙРҪРөР№. Р—РҪР°СҮРөРҪРёСҸ MAE Рё RMSE РҪР°С…РҫРҙСҸСӮСҒСҸ РІ РҙРҫРҝСғСҒСӮРёРјСӢС… РҝСҖРөРҙРөР»Р°С… РҙР»СҸ РҙРҫР»РіРҫСҒСҖРҫСҮРҪСӢС… РәР»РёРјР°СӮРёСҮРөСҒРәРёС… РҝСҖРҫРіРҪРҫР·РҫРІ.

РңРөСӮСҖРёРәР° MAPE РҫРәР°Р·Р°Р»Р°СҒСҢ РҪРөСҒСӮР°РұРёР»СҢРҪРҫР№ РёР·-Р·Р° Р·РҪР°СҮРөРҪРёР№ СӮРөРјРҝРөСҖР°СӮСғСҖСӢ, РұР»РёР·РәРёС… Рә РҪСғР»СҺ, РҝРҫСҚСӮРҫРјСғ РҪРө РёСҒРҝРҫР»СҢР·СғРөСӮСҒСҸ РәР°Рә РҫСҒРҪРҫРІРҪР°СҸ.

Directional accuracy РҪР° СғСҖРҫРІРҪРө 58вҖ“63% РҝРҫРәР°Р·СӢРІР°РөСӮ СҒРҝРҫСҒРҫРұРҪРҫСҒСӮСҢ РјРҫРҙРөР»Рё РәРҫСҖСҖРөРәСӮРҪРҫ РҫРҝСҖРөРҙРөР»СҸСӮСҢ РҪР°РҝСҖР°РІР»РөРҪРёРө РёР·РјРөРҪРөРҪРёСҸ СӮРөРјРҝРөСҖР°СӮСғСҖСӢ.

РҗРҪР°Р»РёР· РҫСҒСӮР°СӮРәРҫРІ (ShapiroвҖ“Wilk, LjungвҖ“Box Рё White test) РҝРҫРәР°Р·Р°Р» РёС… РҪРөРҪРҫСҖРјР°Р»СҢРҪРҫСҒСӮСҢ, РҪР°Р»РёСҮРёРө Р°РІСӮРҫРәРҫСҖСҖРөР»СҸСҶРёРё Рё РіРөСӮРөСҖРҫСҒРәРөРҙР°СҒСӮРёСҮРҪРҫСҒСӮРё, СҮСӮРҫ СҸРІР»СҸРөСӮСҒСҸ РҫР¶РёРҙР°РөРјСӢРј СҖРөР·СғР»СҢСӮР°СӮРҫРј РҙР»СҸ СҒР»РҫР¶РҪСӢС… РҪРөР»РёРҪРөР№РҪСӢС… РІСҖРөРјРөРҪРҪСӢС… СҖСҸРҙРҫРІ СҒ РҙР»РёРҪРҪСӢРј РіРҫСҖРёР·РҫРҪСӮРҫРј РҝСҖРҫРіРҪРҫР·РёСҖРҫРІР°РҪРёСҸ.

```python
# ==============================
# 2.7 РҳРқРўР•Р“Р РҗРҰРҳРҜ Р’ Р•Р”РҳРқР«Рҷ РҹРҗРҷРҹРӣРҗРҷРқ
# ==============================

import numpy as np

# ==============================
# 1. РӣРЈР§РЁРҳР• РңРһР”Р•РӣРҳ
# ==============================

best_models = {
    "Р‘Р»Р°РіРҫРІРөСүРөРҪСҒРә": models["Boosting"],
    "Р“РөР»РөРҪРҙР¶РёРә": models["Boosting"],
    "РңРҫСҒРәРІР°": models["Boosting"],
    "РқР°С…РҫРҙРәР°": models["RandomForest"],
    "РЎР°РҪРәСӮ-РҹРөСӮРөСҖРұСғСҖРі": models["Boosting"],
    "РЎРҫСҮРё": models["Boosting"]
}

# ==============================
# 2. РҹР РҳР—РқРҗРҡРҳ
# ==============================
def make_features(df_city):

    df_city = df_city.sort_values("time").copy()

    for lag in [1, 3, 6, 24, 72, 168]:
        df_city[f"temp_lag_{lag}"] = df_city["temperature_2m"].shift(lag)

    df_city["roll_mean_24"] = df_city["temperature_2m"].rolling(24).mean()
    df_city["roll_std_24"] = df_city["temperature_2m"].rolling(24).std()

    df_city = df_city.dropna()

    return df_city.drop(columns=["city", "time"]).iloc[-1:]


# ==============================
# 3. РҹР РһР“РқРһР— РһР”РқРһР“Рһ Р“РһР РһР”Рҗ
# ==============================
def forecast_city(city):

    model = best_models[city]

    df_city = df_all[df_all["city"] == city].copy()

    X = make_features(df_city)

    pred = model.predict(X)[0]

    return city, pred


# ==============================
# 4. РҹРҗРҷРҹРӣРҗРҷРқ
# ==============================
def full_pipeline():

    results = []

    for city in df_all["city"].unique():
        results.append(forecast_city(city))

    return results


# ==============================
# 5. Р—РҗРҹРЈРЎРҡ
# ==============================
results = full_pipeline()

print("\n===== 2.7 Р Р•Р—РЈРӣР¬РўРҗРў РҹР РһР“РқРһР—Рҗ =====")

for city, pred in results:
    print(f"{city}: {pred:.2f}")
```

*   РјР°СҒСҲСӮР°РұРёСҖРҫРІР°СӮСҢ СҖРөСҲРөРҪРёРө РҪР° РҙСҖСғРіРёРө СҖРөРіРёРҫРҪСӢ
